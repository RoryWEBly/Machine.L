```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
z_values = np.arange(-18, 19)
exp_neg_z = np.exp(-z_values)
one_plus_exp_neg_z = 1 + exp_neg_z
inverse_values = 1 / one_plus_exp_neg_z
data = {
    "z": z_values,
    "exp(-z)": exp_neg_z,
    "1 + exp(-z)": one_plus_exp_neg_z,
    "inverse": inverse_values,
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.scatter(z_values, inverse_values, color='blue', label='Inverse (1 / [1 + exp(-z)])')
plt.title('Scatterplot of Inverse vs z', fontsize=16)
plt.xlabel('z', fontsize=14)
plt.ylabel('Inverse (1 / [1 + exp(-z)])', fontsize=14)
plt.axhline(0.5, color='red', linestyle='--', label='y = 0.5')
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
df.to_excel("scatterplot_data.xlsx", index=False)
```
